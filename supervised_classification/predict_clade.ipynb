```python
import pandas as pd
from collections import defaultdict
import itertools

import sys
sys.path.append("..")
from create_dataset.fasta_to_df import fasta_to_df, all_data_df_to_arange_df

N = 3
BASE = ["A", "T", "G", "C"]
N_BASE = ["".join(i) for i in list(itertools.product(BASE, repeat=N))]
DATA_COLUMNS = ["year", "month", "day"]
CLADE_COLUMNS = ["clade", "head2"]

JAPAN_HEADER_COLUMNS = ["head", "ID", "date"]
ALL_DATA_HEADER_COLUMNS = ["head", "id", "continent", "country", "city", "host", "clade_head", "date"]
```

```python
header, feature = fasta_to_df("all_data_odd_tri", N, ALL_DATA_HEADER_COLUMNS, N_BASE)
```

```python
data = all_data_df_to_arange_df(header, feature)
```

```python
data
```

```python
data = data[data["clade"] != "O"]
data["clade"] = data["clade"].map({"G":0, "GH":1, "GR":2, "GV":3, "L":4, "S":5, "V":6})
data
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
import time
from keras import models
from keras import layers

from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
```

```python
scaler = StandardScaler()

feature = scaler.fit_transform(data[N_BASE])

output_dim = 7
x_train, x_test, y_train, y_test = train_test_split(feature, data["clade"],
                                                    test_size=0.2,
                                                    random_state=42,
                                                    shuffle= True)


# ニューラルネットワークの作成を開始
network = models.Sequential([
    layers.Dense(units=256, activation="relu", input_shape=(feature.shape[1],)),
    layers.Dropout(0.3),
    layers.Dense(units=256, activation="relu"),
    layers.Dropout(0.3),
    layers.Dense(units=256, activation="relu"),
    layers.Dense(units=output_dim, activation="softmax")
])

# ニューラルネットワークをコンパイル
network.compile(loss='sparse_categorical_crossentropy',
                optimizer="adam",
                metrics=["accuracy"])

callbacks = [
    EarlyStopping(patience=3),
    # ModelCheckpoint(model_path, save_best_only=True)
]

# ニューラルネットワークを訓練
hist = network.fit(x_train,
                      y_train,
                      epochs=50,
                      verbose=1,
                      batch_size=128,
                      validation_split = 0.2,
                      callbacks=callbacks)

# テストセットのクラスを予測
# predicted_target = network.predict(features_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score

y_true = y_test
y_pred = network.predict_classes(x_test)

cm = confusion_matrix(y_true, y_pred)
print(accuracy_score(y_true, y_pred))
print(cm)
```
